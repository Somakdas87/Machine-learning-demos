```python
# Run the blocks using shift+enter or click the triangular run button by hovering on the left of the block
# Load the numerical operations packages
import numpy as np
```

```python
# Newton's method with for loop in 1D
# Finding minimizer for y=10*cosh(x/4)-x
xn=0
for i in range(10):
  xnplus1=xn-((5/2)*np.sinh(xn/4)-1)/((5/8)*np.cosh(xn/4))    # x_n+1 = x_n - ( y'(x_n) / y''(x_n) )
  print(xn)   
  xn=xnplus1
```

```python
# Load the graphing package
import matplotlib.pyplot as plt
```

```python
# plot 10*cosh(x/4)-x; Observe that the minimizer is approximately 1.56
x=np.arange(-1,5,0.2)
plt.plot(x,10*np.cosh(x/4)-x)
plt.show()
```

```python
# Gradient descent method; for loop; constant learning rate
xn=0
alpha=1
for i in range(10):
  xnplus1=xn-alpha*((5/2)*np.sinh(xn/4)-1)     # x_n+1 = x_n - alpha*( y'(x_n) )
  print(xn)
  xn=xnplus1
```

```python
# Gradient descent method; for loop; diminishing learning rate
xn=0
alpha=2
for i in range(10):
  xnplus1=xn-(alpha/(i+1))*((5/2)*np.sinh(xn/4)-1)
  print(xn)
  xn=xnplus1
```

```python
# Newton's method with while loop in 1D
epsil=1e-5
err=1
xn=0
while err>epsil:
  xnplus1=xn-((5/2)*np.sinh(xn/4)-1)/((5/8)*np.cosh(xn/4))
  print(xn)
  err=np.abs(xn-xnplus1)
  xn=xnplus1
```

```python
# 3D plot for the following function
def f(x, y):
  return 2*(x-1)**2+3*(y-2)**2

# Steps for plotting

# 1. creating the grid for the domain
x=np.linspace(-5,10,30)
y=np.linspace(-5,10,30)

X,Y=np.meshgrid(x,y)
Z=f(X,Y)

# 2. using grid to plt function
fig=plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# Goal: Verify that the minimizer of z=2(x-1)^2+3*(y-2)^2 is x=1, y=2
# partial derivatives: z_x = 4*(x-1)
#                      z_y = 6*(y-2)
# Gradient descent; for loop; diminishing learning rate

xn=0
yn=0
alpha=0.49
print("{:<20} | {:<20}".format('x', 'y'))
print('-'*45)
for i in range(10):
  xnplus1=xn-(alpha/(i+1))*(4*(xn-1))
  ynplus1=yn-(alpha/(i+1))*(6*(yn-2))
  print("{:<20} | {:<20}".format(xn, yn))
  xn=xnplus1
  yn=ynplus1
```
