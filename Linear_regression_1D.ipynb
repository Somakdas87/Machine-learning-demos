```python
# Linear Regression
import numpy as np                  # Numerical package
import matplotlib.pyplot as plt     # Plotting package
import random                       # Random number generator
from random import random           
```

```python
# Synthesizing a random data set that can fit a linear graph
m=100                               # Size of dataset
x=np.random.normal(0,1,m)+1         # 'm' number of random values in (0,1)
delta=np.random.normal(0,0.8,m)     # introducing chaos
y=2+delta+(1.5)*x                   # imitating the line y= 1.5x + 2 with error
```

```python
# Plotting synthesized data set
plt.scatter(x,y)
```

```python
# Gradient Descent
theta0=0
theta1=0.1
alpha=0.2
print("{:>30}        | {:>14}".format('theta_0','theta_1'))
print('-'*58)
for i in range(1001):
  newTheta0=theta0-alpha*((1/m)*np.sum(theta0+theta1*x-y))      # Component 1
  newTheta1=theta1-alpha*((1/m)*np.sum((theta0+theta1*x-y)*x))  # Component 2
  if i%50==0:
    print("Iteration {:<5}: {:<20} | {:<20}".format(i,theta0, theta1))
  theta0=newTheta0
  theta1=newTheta1
```

```python
# Plot the line with the above parametric values to check that it fits the dataset
plt.scatter(x,y)
plt.plot(x, theta0+theta1*x,color='red')
```

```python
# Gradient Descent
theta0=0
theta1=0.1
alpha=0.2
fig,ax=plt.subplots()
for i in range(1000):
  newTheta0=theta0-alpha*((1/m)*np.sum(theta0+theta1*x-y))
  newTheta1=theta1-alpha*((1/m)*np.sum((theta0+theta1*x-y)*x))
  t=np.arange(-1,4)
  # plotting the line for the first 10 steps showing the evolution
  if i in range(10):
    ax.plot(t, theta0+theta1*t, label= 'Iter'+str(i)) 
  theta0=newTheta0
  theta1=newTheta1
legend=ax.legend()
ax.scatter(x,y)
plt.show()
```
